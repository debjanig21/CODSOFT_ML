```python
X=data.drop('is_fraud',axis=1)
y=data['is_fraud']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import OrdinalEncoder
cols =['trans_date_trans_time','street','full name','city','state','job','dob','trans_num']
encoder =OrdinalEncoder()
X[cols]=encoder.fit_transform(X[cols])
```

```python
data.columns
```

```python
from imblearn.over_sampling import SMOTE
X_res,y_res=SMOTE().fit_resample(X,y)
X_res.value_counts()
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.20,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
y_pred1=log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred2=dt.predict(X_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
precision_score(y_test,y_pred2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred3=rf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred3)
```

```python
precision_score(y_test,y_pred3)
```

```python
final_data=pd.DataFrame({'Models':['LR','DT','RF'],'ACC':[accuracy_score(y_test,y_pred1),accuracy_score(y_test,y_pred2),accuracy_score(y_test,y_pred3)]})
```
